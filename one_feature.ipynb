```python
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the model function
def model(x, w1, w2, beta):
    return w1 * x**w2 + beta
```

```python
# Generate 1000 random numbers for x_data
x_data = np.random.rand(1000)

# true parameters
true_params = [2, 0.3, -4]

# Calculate y_data using the formula y = 2*x^0.3 - 4
y_data = model(x_data, *true_params) + np.random.normal(0, 0.2, x_data.size)
```

```python
# Fit the model to the data
popt, _ = curve_fit(model, x_data, y_data, p0=[0, 0, 1])
```

```python
# Extract the parameters
w1, w2, beta = popt

# Generate x values for the fitted curve
x_fit = np.linspace(0, 1, 100)
y_fit = model(x_fit, *popt)

# Create the formula string with LaTeX formatting for the legend
fitted_formula = f'$y = {w1:.2f} x^{{{w2:.2f}}} + {beta:.2f}$'
true_formula = f'$y = {true_params[0]:.2f} x^{{{true_params[1]:.2f}}} + {true_params[2]:.2f}$'

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the data points
plt.scatter(x_data, y_data, color='blue', alpha=0.5, label=f'Data points:\t {true_formula}')

# Plot the fitted curve
plt.plot(x_fit, y_fit, color='red', label=f'Fitted curve:\t {fitted_formula}')

# Add labels and title
plt.xlabel('x_data')
plt.ylabel('y_data')
plt.title('Data Points and Fitted Curve')

# Add the legend with the formula included
plt.legend(loc='best')

# save plot
plt.savefig("figs/one_feature.png")

# Show the plot
plt.show()
```
